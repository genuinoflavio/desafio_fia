```python
# mostre na tela 10 pokemons e sua lista de movimentos
join_cond = (dfPokemon.type == dfTypes.name)
(
  dfPokemon.alias('Pokemon')
  .join(dfTypes.alias('Types'), join_cond, how='inner')
  .select('Pokemon.nome','Types.moves')
  .limit(10)
  .display()
)
```

```python
# mostre um explode SOMENTE dos pokemons que possuem mais de uma forma
(
  dfPokemon
  .filter(fn.size('formas') > 1)
  .select('nome',fn.explode_outer('formas'))
  .display()
)
```

```python
# mostre o nome dos pokemons e uma lista simples de seus movimentos]
(
  dfPokemon
  .join(dfTypes.alias('Types'), (dfPokemon.type == dfTypes.name), how='inner')
  .select('nome',fn.explode_outer('moves.name').alias('move'))
  .groupBy('nome')
  .agg(fn.collect_list('move').alias('lista_moves'))
  .display()
)
```

```python
# dos pokemons do tipo 'rock', quantos movimentos possuem os pokemons mais pesados?
(
  dfPokemon
  .filter( (fn.col('type') == 'rock') )
  .filter(fn.col('peso') == dfPokemon.filter((fn.col('type') == 'rock')).agg(fn.max('peso')).collect()[0][0])
  .join(dfTypes.alias('Types'), (dfPokemon.type == dfTypes.name), how='inner')
  .select('nome',fn.explode_outer('moves'))
  .groupBy('nome')
  .agg(fn.collect_list('col.name').alias('lista_moves'))
  .withColumn('qtd_movimentos', fn.size(fn.col('lista_moves')))
  .display()
)
```

```python
# considerando que pokemons leves estão abaixo de 1000kg, mostre quantos pokemos leves e pesados existem em cada categoria de peso
(
  dfPokemon
  .withColumn('classe_peso',
             fn.when(fn.col('peso') < 1000, fn.lit('leve'))
              .otherwise(fn.lit('pesado'))
             )
  .groupBy('classe_peso')
  .agg(fn.count(fn.lit(1)))
  .display()
)
```
